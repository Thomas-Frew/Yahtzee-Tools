```python
import pandas as pd
```

```python
# Load dataset and eplace column names
dataset_name = "input_dataset.csv"
column_names = ["Ones", "Twos", "Threes", "Fours", "Fives", "Sixes", "Total", "Bonus", "3 of a Kind", "4 of a Kind", "Full House", "Small Straight", "Large Straight", "Chance", "Yahtzee", "Score"]

df = pd.read_csv(dataset_name, dtype=str, keep_default_na=False, names=column_names)
df
```

```python
# Replace player IDs with unique nonces
df.index = pd.factorize(df.index)[0]
df
```

```python
# Export the dataset
df.to_csv('output_dataset.csv')
```
